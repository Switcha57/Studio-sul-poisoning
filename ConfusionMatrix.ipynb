# Confusion Matrix Plotter

This notebook allows you to input True Negative (TN), False Positive (FP), False Negative (FN), and True Positive (TP) values to generate and save a confusion matrix image.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plot
sns.set_style("white")
```

```python
# Define the values for the confusion matrix
# Replace these values with your specific results
TN = 1588  # True Negatives
FP = 412  # False Positives
FN = 259    # False Negatives
TP = 1741  # True Positives

output_filename = "DnnEmTrainSet.png"
```

```python
def plot_confusion_matrix(tn, fp, fn, tp, output_file):
    # Construct the confusion matrix
    cm = np.array([[tn, fp], [fn, tp]])
    
    plt.figure(figsize=(8, 6))
    
    # Labels for the quadrants
    group_names = ['True Neg','False Pos','False Neg','True Pos']
    group_counts = ["{0:0.0f}".format(value) for value in cm.flatten()]
    
    # # Calculate percentages
    # if np.sum(cm) > 0:
    #     group_percentages = ["{0:.2%}".format(value) for value in cm.flatten()/np.sum(cm)]
    # else:
    #     group_percentages = ["0.00%" for _ in cm.flatten()]
    
    labels = [f"{v1}" for v1 in group_counts]
    labels = np.asarray(labels).reshape(2,2)
    
    sns.heatmap(cm, annot=labels, fmt='', cmap='Blues', cbar=False,
                xticklabels=['Goodware', 'Malware'],
                yticklabels=['Goodware', 'Malware'],
                annot_kws={"size": 24})
    
    plt.ylabel('Reale')
    plt.xlabel('Predetto')
    plt.tight_layout()
    
    plt.savefig(output_file)
    print(f"Confusion matrix saved to {output_file}")
    plt.show()

# Run the function
plot_confusion_matrix(TN, FP, FN, TP, output_filename)
```
